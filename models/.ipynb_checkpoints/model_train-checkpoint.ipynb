```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler, MinMaxScaler 
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import SVC 
from xgboost import XGBClassifier 
from sklearn import metrics 

import tensorflow as tf

import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import r2_score, mean_squared_error
  
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../data/TSLA.csv")
```

```python
df.drop(df.columns[0], inplace=True, axis=1)
```

```python
split = df['Date'].str.split('-', expand=True) 
  
df['day'] = split[2].astype('int') 
df['month'] = split[1].astype('int') 
df['year'] = split[0].astype('int') 
  
df.head()
```

```python
df['is_quarter_end'] = np.where(df['month']%3==0, 1, 0) 
df.head(10)
```

```python
df.groupby('is_quarter_end').describe()
```

```python
df['open-close']  = df['Open'] - df['Close'] 
df['low-high']  = df['Low'] - df['High'] 
df['target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

```python
features = df[['open-close', 'low-high', 'is_quarter_end']] 
target = df['target'] 
  
scaler = StandardScaler() 
features = scaler.fit_transform(features) 
  
X_train, X_valid, Y_train, Y_valid = train_test_split( 
    features, target, test_size=0.2, random_state=0) 
print(X_train.shape, X_valid.shape) 
```

```python
clf = LogisticRegression().fit(X_train, Y_train) 
  
print('Logistic Regression: ') 
print('Training Accuracy : ', metrics.roc_auc_score( 
    Y_train, clf.predict_proba(X_train)[:,1])) 
print('Validation Accuracy : ', metrics.roc_auc_score( 
    Y_valid, clf.predict_proba(X_valid)[:,1])) 
print('Mean Accuracy: ', clf.score(X_train, Y_train))
```
